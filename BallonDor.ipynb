```python
import matplotlib.pyplot as plt
```

```python
players = ["Lionel Messi", "Cristiano Ronaldo", "Michel Platini", "Johan Cruyff",
           "Marco van Basten", "Franz Beckenbauer", "Ronaldo Naz√°rio", "Di Stefano",
           "Kevin Keegan", "Rummenigge"]
wins = [8, 5, 3, 3, 3, 2, 2, 2, 2, 2]
```

```python
plt.figure(figsize=(12,6))
plt.barh(players[::-1], wins[::-1], color='gold')
plt.xlabel("Number of Ballon d'Or Wins")
plt.title("Top 10 Players with Most Ballon d'Or Wins")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
