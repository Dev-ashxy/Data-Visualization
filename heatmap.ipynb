```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.rand(10, 10)
df = pd.DataFrame(data, columns=[f'Col{i}' for i in range(1, 11)])
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='Reds', cbar=True)
plt.title('Heatmap of Random Dataset')
plt.show()
```
