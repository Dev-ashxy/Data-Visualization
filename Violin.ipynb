```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create sample data for different machine learning models' scores
np.random.seed(42)
data = {
    'Model': np.repeat(['Model A', 'Model B', 'Model C'], 100),
    'Accuracy': np.concatenate([
        np.random.normal(loc=0.88, scale=0.03, size=100), # Model A is consistent
        np.random.normal(loc=0.92, scale=0.08, size=100), # Model B has high variance
        np.append(np.random.normal(0.80, 0.02, 50), np.random.normal(0.95, 0.02, 50)) # Model C is bimodal
    ])
}
df = pd.DataFrame(data)

# Create the violin plot
plt.figure(figsize=(12, 7))
sns.violinplot(x='Model', y='Accuracy', data=df, palette='muted', inner='quartile')
# inner='quartile' draws lines for the quartiles inside the violins

plt.title('Comparison of Model Accuracy Distributions')
plt.xlabel('Machine Learning Model')
plt.ylabel('Accuracy Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
