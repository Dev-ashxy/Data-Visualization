```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Average daily time spent on social media (in minutes)
platforms = ['TikTok', 'Instagram', 'X (Twitter)', 'Facebook', 'LinkedIn']
time_minutes = [95, 78, 52, 38, 17]
```

```python
# Convert minutes to hours for better readability
time_hours = [minutes / 60 for minutes in time_minutes]
```

```python
# Create color palette
colors = ['#ff0050', '#E4405F', '#1DA1F2', '#1877F2', '#0A66C2']
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Create horizontal bar chart
bars = ax.barh(platforms, time_hours, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

# Add value labels on the bars
for i, (bar, minutes) in enumerate(zip(bars, time_minutes)):
    width = bar.get_width()
    hours = minutes // 60
    mins = minutes % 60
    
    # Format time string
    if hours > 0:
        time_label = f'{hours}h {mins}m'
    else:
        time_label = f'{mins}m'
    
    # Add text annotation on the bar
    ax.text(width + 0.05, bar.get_y() + bar.get_height()/2, 
            time_label, 
            va='center', ha='left', fontsize=12, fontweight='bold')

# Customize the chart
ax.set_xlabel('Average Daily Time (hours)', fontsize=14, fontweight='bold')
ax.set_ylabel('Social Media Platform', fontsize=14, fontweight='bold')
ax.set_title('Daily Social Media Usage by Platform\nAverage Time Spent (2024)', 
             fontsize=16, fontweight='bold', pad=20)

# Add grid for better readability
ax.grid(axis='x', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

# Set x-axis limits
ax.set_xlim(0, max(time_hours) + 0.5)

# Add total time annotation
total_minutes = sum(time_minutes)
total_hours = total_minutes // 60
total_mins = total_minutes % 60

# Add a text box with total time
textstr = f'Total Daily Time:\n{total_hours}h {total_mins}m'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax.text(0.98, 0.97, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', horizontalalignment='right', bbox=props)

# Add insights as annotations
insights = [
    'Most engaging platform',
    'Visual content dominates',
    'Quick news updates',
    'Declining usage',
    'Professional networking'
]

for i, (platform, insight) in enumerate(zip(platforms, insights)):
    ax.annotate(insight, 
                xy=(time_hours[i], i), 
                xytext=(time_hours[i] + 0.3, i),
                fontsize=9, 
                style='italic',
                color='gray')

plt.tight_layout()
plt.show()
```

```python
# Print summary statistics
print("\n=== Social Media Usage Summary ===")
print(f"Total platforms analyzed: {len(platforms)}")
print(f"Total daily time: {total_hours}h {total_mins}m ({total_minutes} minutes)")
print(f"Average time per platform: {total_minutes // len(platforms)} minutes")
print(f"Most used platform: {platforms[0]} ({time_minutes[0]} minutes)")
print(f"Least used platform: {platforms[-1]} ({time_minutes[-1]} minutes)")
```
