```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the plot style
plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(10, 8))

# Generate the data and plot the lines
for i in range(1, 51):
    y_offset = i * 0.8
    
    x = np.linspace(0, 10, 200)
    y_base = np.exp(-((x - 5)**2) / 2)
    y_noise = np.random.normal(0, 0.03, x.shape[0]) * (np.sin(x * 2) + 1.2)
    y = (y_base + y_noise) * 5 + y_offset

    
    ax.plot(x, y, color='white', lw=1.5)
    ax.fill_between(x, y_offset, y, color='black', zorder=i)


ax.set_title('"Unknown Signals"', fontsize=20, color='white')
ax.axis('off') 
plt.show()
```
