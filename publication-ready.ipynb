```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Generate sample data
np.random.seed(42)
x = np.linspace(0, 10, 200)
y = np.sin(x) * np.exp(-0.1 * x)
scatter_x = np.random.rand(50) * 10
scatter_y = np.random.rand(50) * 0.6 - 0.3
hist_data = y + np.random.normal(0, 0.05, 200)

# 1. Create the figure
fig = plt.figure(figsize=(14, 7))
fig.suptitle('Advanced Financial Analysis Dashboard', fontsize=18)

# 2. Define the grid layout
gs = gridspec.GridSpec(2, 3, figure=fig)

ax_main = fig.add_subplot(gs[:, 0:2]) # Main plot spans all rows, first two columns
ax_hist = fig.add_subplot(gs[0, 2])   # Histogram on row 0, col 2
ax_scatter = fig.add_subplot(gs[1, 2]) # Scatter on row 1, col 2

# 3. Populate the plots
# Main plot
ax_main.plot(x, y, color='navy', label='Damped Sine Wave')
ax_main.set_title('Main Trend Analysis')
ax_main.set_xlabel('Time')
ax_main.set_ylabel('Value')
ax_main.grid(True, linestyle=':')
ax_main.axhline(0, color='black', linewidth=0.5)

# Add a significant event annotation
peak_x, peak_y = 4.75, -0.27
ax_main.annotate('Critical Event', xy=(peak_x, peak_y),
                 xytext=(peak_x + 1, peak_y + 0.3),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
                 fontsize=12, backgroundcolor='yellow')

# Histogram
ax_hist.hist(hist_data, bins=30, color='darkorange', orientation='horizontal')
ax_hist.set_title('Value Distribution')

# Scatter plot
ax_scatter.scatter(scatter_x, scatter_y, color='green', alpha=0.6)
ax_scatter.set_title('Residuals')

# 4. Final layout adjustments
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
